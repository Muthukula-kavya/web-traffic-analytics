```python
import pandas as pd

df = pd.read_csv("global_web_traffic_2026.csv")
print(df.head())
print(df.columns)
```

```python
print("Dataset Shape:", df.shape)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
top_sites = df.sort_values("monthly_visits", ascending=False).head(10)

print(top_sites[["domain", "monthly_visits"]])
```

```python
import matplotlib.pyplot as plt

top_sites = df.sort_values("monthly_visits", ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top_sites["domain"], top_sites["monthly_visits"])
plt.xticks(rotation=45)
plt.title("Top 10 Websites by Monthly Visits")
plt.tight_layout()
plt.show()
```

```python
print("Average Bounce Rate:")
print(df["bounce_rate_pct"].mean())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["bounce_rate_pct"], bins=10)
plt.title("Bounce Rate Distribution")
plt.xlabel("Bounce Rate (%)")
plt.ylabel("Count")
plt.show()
```

```python
print("Average Session Duration:")
print(df["avg_session_duration_s"].mean())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["avg_session_duration_s"], bins=10)
plt.title("Session Duration Distribution")
plt.xlabel("Duration (Seconds)")
plt.ylabel("Count")
plt.show()
```

```python
print(df["category"].value_counts())
```

```python
df["category"].value_counts().head(10).plot(kind="bar")

plt.title("Top Website Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
print(df["primary_market"].value_counts().head(10))
```
